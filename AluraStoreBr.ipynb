### Importação dos dados



```python
import pandas as pd
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja1 = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja1.head()
```

#1. Análise do faturamento


```python
# Lista com nomes das lojas
nomes_lojas = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']

# Calcula o faturamento total de cada loja
faturamentos_totais = [
    (loja1['Preço'] + loja1['Frete']).sum(),
    (loja2['Preço'] + loja2['Frete']).sum(),
    (loja3['Preço'] + loja3['Frete']).sum(),
    (loja4['Preço'] + loja4['Frete']).sum()
]

# Mostra no terminal
for i, valor in enumerate(faturamentos_totais):
    print(f'O faturamento da {nomes_lojas[i]} foi de R$ {valor:.2f}')

```

# 2. Vendas por Categoria


```python
vendas_por_categoria_loja1 = loja1['Categoria do Produto'].value_counts()
vendas_por_categoria_loja2 = loja2['Categoria do Produto'].value_counts()
vendas_por_categoria_loja3 = loja3['Categoria do Produto'].value_counts()
vendas_por_categoria_loja4 = loja4['Categoria do Produto'].value_counts()

# Exibir os resultados
print("Loja 1:\n", vendas_por_categoria_loja1)
print("\nLoja 2:\n", vendas_por_categoria_loja2)
print("\nLoja 3:\n", vendas_por_categoria_loja3)
print("\nLoja 4:\n", vendas_por_categoria_loja4)

```

# 3. Média de Avaliação das Lojas

```python
loja1_avaliacao = loja1['Avaliação da compra'].mean()
loja2_avaliacao = loja2['Avaliação da compra'].mean()
loja3_avaliacao = loja3['Avaliação da compra'].mean()
loja4_avaliacao = loja4['Avaliação da compra'].mean()

avaliacoes = [loja1_avaliacao, loja2_avaliacao, loja3_avaliacao, loja4_avaliacao]

print(f'A avaliação média da loja 1 foi de {loja1_avaliacao:.2f}')
print(f'A avaliação média da loja 2 foi de {loja2_avaliacao:.2f}')
print(f'A avaliação média da loja 3 foi de {loja3_avaliacao:.2f}')
print(f'A avaliação média da loja 4 foi de {loja4_avaliacao:.2f}')
```

# 4. Produtos Mais e Menos Vendidos

```python
# Contagem de cada produto vendido
produtos_vendidos_loja1 = loja1['Produto'].value_counts()
produtos_vendidos_loja2 = loja2['Produto'].value_counts()
produtos_vendidos_loja3 = loja3['Produto'].value_counts()
produtos_vendidos_loja4 = loja4['Produto'].value_counts()

# Exibir os produtos mais vendidos e menos vendidos para cada loja
print("Loja 1 - Produto Mais Vendido:", produtos_vendidos_loja1.idxmax(), "com", produtos_vendidos_loja1.max(), "vendas")
print("Loja 1 - Produto Menos Vendido:", produtos_vendidos_loja1.idxmin(), "com", produtos_vendidos_loja1.min(), "vendas")

print("\nLoja 2 - Produto Mais Vendido:", produtos_vendidos_loja2.idxmax(), "com", produtos_vendidos_loja2.max(), "vendas")
print("Loja 2 - Produto Menos Vendido:", produtos_vendidos_loja2.idxmin(), "com", produtos_vendidos_loja2.min(), "vendas")

print("\nLoja 3 - Produto Mais Vendido:", produtos_vendidos_loja3.idxmax(), "com", produtos_vendidos_loja3.max(), "vendas")
print("Loja 3 - Produto Menos Vendido:", produtos_vendidos_loja3.idxmin(), "com", produtos_vendidos_loja3.min(), "vendas")

print("\nLoja 4 - Produto Mais Vendido:", produtos_vendidos_loja4.idxmax(), "com", produtos_vendidos_loja4.max(), "vendas")
print("Loja 4 - Produto Menos Vendido:", produtos_vendidos_loja4.idxmin(), "com", produtos_vendidos_loja4.min(), "vendas")
```

# 5. Frete Médio por Loja


```python
frete_medio1 = loja1['Frete'].mean();
frete_medio2 = loja2['Frete'].mean();
frete_medio3 = loja3['Frete'].mean();
frete_medio4 = loja4['Frete'].mean();

fretes = [frete_medio1, frete_medio2, frete_medio3, frete_medio4]
print(f'O frete médio da loja 1 foi de R$ {frete_medio1:.2f}')
print(f'O frete médio da loja 2 foi de R$ {frete_medio2:.2f}')
print(f'O frete médio da loja 3 foi de R$ {frete_medio3:.2f}')
print(f'O frete médio da loja 4 foi de R$ {frete_medio4:.2f}')
```

Relatório

Como a média das avaliações e a categoria de vendas por loja não apresentam diferença relevante. A análise torna possível crer que o faturamento é o maior indicador de que a loja 4 é a que tem a menor performance e portanto deve ser desligada.


```python
# Lista de vendas por categoria para cada loja
vendas_por_categoria = [vendas_por_categoria_loja1, vendas_por_categoria_loja2,
                        vendas_por_categoria_loja3, vendas_por_categoria_loja4]

# Cores diferentes para cada categoria
cores = plt.cm.Paired(range(len(vendas_por_categoria_loja1)))

# Criar gráficos de pizza para cada loja
fig, axes = plt.subplots(2, 2, figsize=(10, 10))  # Cria uma matriz 2x2 de subgráficos

for i, ax in enumerate(axes.flatten()):
    # Dados de vendas por categoria de cada loja
    dados = vendas_por_categoria[i]

    # Gráfico de pizza para as vendas por categoria e % de venda
    ax.pie(dados, labels=dados.index, autopct='%1.1f%%', colors=cores, startangle=90)
    ax.set_title(f'Distribuição por Categoria - {nomes_lojas[i]}')

plt.tight_layout()  # Ajuste para que os gráficos não se sobreponham
plt.show()
# Gráfico de barras Faturamento
plt.bar(nomes_lojas, faturamentos_totais, color = ['#FF6347', '#4682B4', '#32CD32', '#FFD700'] )
plt.title('Faturamento por Loja')
plt.ylabel('R$')
plt.xlabel('Loja')
plt.tight_layout()
plt.show()

#Gráfico de barras avaliacao media
plt.barh(nomes_lojas, avaliacoes, color = ['#FF6347', '#4682B4', '#32CD32', '#FFD700'] )
plt.title('Avaliação média por Loja')
plt.xlabel('R$')
plt.ylabel('Loja')
plt.tight_layout()
plt.show()
```
